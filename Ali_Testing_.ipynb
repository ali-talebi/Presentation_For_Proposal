```python
### N(1 , 0.25 )

mean_delta_x = (0.05 , 0.02 )
mean_delta_y = (0.05 , 0.03)
sample_total_data = []
for i in range(100) :
  sub_run = []

  for element in data :

    r = np.random.randint(-1 , 2 )
    z_alpha_x = np.random.rand()
    z_alpha_y = np.random.rand()
    x_new = element[0] + r * gauss( mean_delta_x[0]  ,  (  mean_delta_x[1] / ( 100** 0.5 ) ) )
    y_new = element[1] + r * gauss( mean_delta_y[0]  ,  (  mean_delta_y[1] / ( 100** 0.5 ) ) )
    sub_run.append([x_new , y_new ])



  sub_run = np.array(sub_run)
  sample_total_data.append(sub_run)














```
