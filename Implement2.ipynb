```python
underform_balde = np.array([[1 , 1  , 1 ],
                            [1 , 2  , 1 ],
                            [2 , 1  , 1 ],
                            [2 , 2  , 1 ],
                            [3 , 1  , 1 ],
                            [3 , 2  , 1 ],
                            [4 , 1  , 1 ],
                            [4 , 2  , 1 ]
                            ])
```

```python
derform_balde = np.array(  [[1+0.01 , 1+0.05  , 1 ],
                            [1+0.01 , 2+0.05  , 1 ],
                            [2+0.01 , 1+0.1  , 1+0.002 ],
                            [2+0.01 , 2+0.1  , 1+0.002 ],
                            [3+0.03 , 1+0.15  , 1+0.005 ],
                            [3+0.03 , 2+0.15  , 1+0.005 ],
                            [4+0.04 , 1+0.2  , 1+0.006 ],
                            [4+0.04 , 2+0.2  , 1+0.006 ]
                            ])
```

```python
x_vals = underform_balde[ : , 0 ]
y_vals = underform_balde[ : , 1 ]
z_vals = underform_balde[ : , 2 ]
x_vals_deform = derform_balde[ : , 0 ]
y_vals_deform = derform_balde[ : , 1 ]
z_vals_deform = derform_balde[ : , 2 ]
fig = plt.figure(figsize = (5 , 5 ))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_vals, y_vals, z_vals, c = 'b', marker='o')
ax.scatter(x_vals_deform, y_vals_deform, z_vals_deform, s=30 , c = 'r', marker='*')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

```python
x_vals_deform = derform_balde[ : , 0 ]
y_vals_deform = derform_balde[ : , 1 ]
z_vals_deform = derform_balde[ : , 2 ]
fig = plt.figure(figsize = (5 , 5 ))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_vals, y_vals, z_vals, c = 'b', marker='o')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```
